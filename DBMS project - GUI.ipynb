# Final GUI with login

```python
import tkinter as tk                    
from tkinter import ttk
from tkinter import *
import sqlite3
from tkinter import messagebox as msgbox
from datetime import date
import time
  
#main window to contain the tabs (login and register)  
root = tk.Tk()
root.title("User Window")
root.geometry("500x500")

#notebook is a widget that allows to switch between tabs
tabControl = ttk.Notebook(root)
 
#creating the tabs
tab1 = ttk.Frame(tabControl)
tab2 = ttk.Frame(tabControl)

#naming the tabs
tabControl.add(tab1, text ='Login')
tabControl.add(tab2, text ='Register')

#positioning the tabs
tabControl.pack(expand = 1, fill ="both")

"""Tab 1 ---------  for login"""

#function to change background colour
def con(a):
    a.config(bg="grey")
def b(j):
    j.config(bg="white")

#label for username
user=Label(tab1,text="Username",font=('Merriweather',13))
user.place(relx=0.3,rely=0.3,anchor=CENTER)
con(user)

#label for password
pw=Label(tab1,text="Password",font=('Merriweather',13))
pw.place(relx=0.3,rely=0.38,anchor=CENTER)
con(pw)

#entry widget for username and password
u=Entry(tab1,width=50,font=('Comic Sans MS',13))
u.place(relx=0.62,rely=0.3,anchor=CENTER,width=175)
b(u)
p=Entry(tab1,show="*",width=50,font=('Comic Sans MS',13))
p.place(relx=0.62,rely=0.38,anchor=CENTER,width=175)
b(p)


"""Login function"""
def logintodb():
    #getting data
    uname=u.get()
    pwd=p.get()
    #applying empty validation
    if uname=='' or pwd=='':
        msg4 = msgbox.showinfo("Status","Fill details!!")

    else:
        #open database
        curr = sqlite3.connect('user.db')
        conn = curr.cursor() 
        #select query
        cursor = conn.execute('SELECT * from USER where USERNAME="%s" and PASSWORD="%s"'%(uname,pwd))
        
        #fetch data 
        if cursor.fetchone():
            msg5 = msgbox.showinfo("Status","Login successful!!")
            
            """seperate window for admin"""
            if uname=='admin' and pwd=='admin':
                admin_window=Tk()
                admin_window.title("User details")
                admin_window.geometry('500x500')
                admin_window.config(bg='black')
                
                #display user details
                r_set=conn.execute('''SELECT * from USER_MESSAGE''');
                i=0 # row value inside the loop 
                for student in r_set: 
                    for j in range(len(student)):
                        e = Entry(admin_window, width=10, fg='blue') 
                        e.grid(row=i, column=j) 
                        e.insert(END, student[j])
                    i=i+1
                
                
                """update window"""
                #update function
                def Update():
                    upd_window = Tk()
                    upd_window.title("Enter new details")
                    upd_window.geometry("500x500")
                    
                    #label for old username
                    old_uname=Label(upd_window,text="Username",font=('Merriweather',13))
                    old_uname.place(relx=0.3,rely=0.3,anchor=CENTER)
                    con(old_uname)
                    
                    #label for new username
                    new_uname=Label(upd_window,text="New Username",font=('Merriweather',13))
                    new_uname.place(relx=0.3,rely=0.38,anchor=CENTER)
                    con(new_uname)
                    
                    #label for new password
                    new_pw=Label(upd_window,text="Password",font=('Merriweather',13))
                    new_pw.place(relx=0.3,rely=0.46,anchor=CENTER)
                    con(new_pw)

                    #entry widget for old username
                    old_u=Entry(upd_window,width=50,font=('Comic Sans MS',13))
                    old_u.place(relx=0.62,rely=0.3,anchor=CENTER,width=175)
                    b(old_u)
                    
                    #entry widget for new username
                    new_u=Entry(upd_window,width=50,font=('Comic Sans MS',13))
                    new_u.place(relx=0.62,rely=0.38,anchor=CENTER,width=175)
                    b(new_u)
                    
                    #entry widget for new password
                    new_p=Entry(upd_window,show="*",width=50,font=('Comic Sans MS',13))
                    new_p.place(relx=0.62,rely=0.46,anchor=CENTER,width=175)
                    b(new_p)

                    #storing the details 
                    old_username = old_u.get()
                    new_username = new_u.get()
                    new_password = new_p.get()
                    
                    def Save():
                        conn.execute('UPDATE USER SET USERNAME="%s",PASSWORD="%s" WHERE USERNAME="%s"'%(new_username,new_password,old_username))
                        curr.commit()
                        conn.execute('UPDATE USER_MESSAGE SET USERNAME="%s" where USERNAME="%s"'%(new_username,old_username))
                        curr.commit()
                        msg= msgbox.showinfo("Status","Successfully Updated!!!")

                    #save button
                    save = Button(upd_window, text = "Save", command = Save)
                    save.place(relx=0.5,rely=0.54)
            
                #update button widget
                update = Button(admin_window, text = "Update", command = Update)
                update.grid(row=i+1, column=1,padx=10,pady=10)
                
                
                """delete button"""
                def Delete():
                    del_window = Tk()
                    del_window.title("Enter details to be deleted")
                    del_window.geometry("500x500")
                    
                    #label for delete username
                    del_uname=Label(del_window,text="Username",font=('Merriweather',13))
                    del_uname.place(relx=0.3,rely=0.3,anchor=CENTER)
                    con(del_uname)

                    #entry widget for delete username
                    del_u=Entry(del_window,width=50,font=('Comic Sans MS',13))
                    del_u.place(relx=0.62,rely=0.3,anchor=CENTER,width=175)
                    b(del_u)
                    
                    #storing details
                    username = del_u.get()
                    
                    def Remove():
                        conn.execute('DELETE FROM USER where USERNAME="%s"'%(username))
                        curr.commit()
                        conn.execute('DELETE FROM USER_MESSAGE where USERNAME="%s"'%(username))
                        curr.commit()
                        msg= msgbox.showinfo("Status","Records successfully Deleted!!!")
                        
                    #remove button
                    remove = Button(del_window, text = "Remove user", command = Remove)
                    remove.place(relx=0.5,rely=0.34)
                
                #delete button widget
                delete = Button(admin_window, text = "Delete", command = Delete)
                delete.grid(row=i+1, column=2,padx=10,pady=10)
                
                
            else:
                """seperate window for remaining users"""
                window=Tk()
                window.title("Home page")
                window.geometry('500x500')
                window.config(bg='black')
                
                #label for user message
                message=Label(window,text="Message",font=('Merriweather',13))
                message.place(relx=0.3,rely=0.38,anchor=CENTER)
                con(message)

                #text area to write user message
                m=Text(window,font=('Comic Sans MS',13))
                m.place(relx=0.62,rely=0.4,anchor=CENTER,height=50,width=175)
                b(m)
        
        
                #send message function for home page
                def send():
                    
                    #username , message and datetime of sending message
                    uname = u.get() 
                    mess = m.get("1.0","end-1c")
                    d = date.today()
                    t = time.localtime()
                    current_time = time.strftime("%H:%M", t)
                
                    #connect to db
                    db=sqlite3.connect("user.db")
                    conn=db.cursor()
                    
                    #message table
                    conn.execute("""CREATE TABLE IF NOT EXISTS USER_MESSAGE(USERNAME TEXT,M_DATE DATE, M_TIME TIME, MESSAGE TEXT NOT NULL);""")
                    conn.execute("""INSERT INTO USER_MESSAGE(USERNAME,M_DATE,M_TIME,MESSAGE) VALUES(?,?,?,?);""",(uname,d,current_time,mess))
                    
                    #alert if message saved
                    msg3 = msgbox.showinfo( "Status"," Message saved ")
                    
                    cur=conn.execute("SELECT * FROM USER_MESSAGE")
                    print("Details of user messages : ")
                    for row in cur:
                        print(row)
                    print()
                    
                    db.commit()
                    db.close()
                
                #send button
                send=Button(window,text="Send",font=('Merriweather',13),command=send)
                send.place(relx=0.72,rely=0.5,anchor=CENTER)
                send.config(bg="#f4d160")

                window.mainloop()
            
        #alert for wrong credentials
        else:
            msg4 = msgbox.showinfo("Status","Wrong credentials")
    

#login button    
log=Button(tab1,text="Login",font=('Merriweather',13),command=logintodb)
log.place(relx=0.735,rely=0.475,anchor=CENTER)
log.config(bg="#f4d160")




"""Tab 2  -------- for registration"""

#label for username
user_n=Label(tab2,text="Username",font=('Merriweather',13))
user_n.place(relx=0.3,rely=0.3,anchor=CENTER)
con(user_n)

#label for password
pw_n=Label(tab2,text="Password",font=('Merriweather',13))
pw_n.place(relx=0.3,rely=0.38,anchor=CENTER)
con(pw_n)

#entry widget for username and password
u_n=Entry(tab2,width=50,font=('Comic Sans MS',13))
u_n.place(relx=0.62,rely=0.3,anchor=CENTER,width=175)
b(u_n)
p_n=Entry(tab2,show="*",width=50,font=('Comic Sans MS',13))
p_n.place(relx=0.62,rely=0.38,anchor=CENTER,width=175)
b(p_n)



#function to add new user to db
def register():
    u = u_n.get()
    p = p_n.get()
    db=sqlite3.connect("user.db")
    conn=db.cursor()
    conn.execute("""CREATE TABLE IF NOT EXISTS USER(USERNAME TEXT NOT NULL PRIMARY KEY,PASSWORD TEXT NOT NULL);""")
    
    #check if username already exists
    conn.execute("SELECT USERNAME FROM USER WHERE USERNAME = ?", (u,))
    data=conn.fetchall()
    
    if len(data)==0:#if not add new user
        conn.execute("""INSERT INTO USER(USERNAME,PASSWORD) VALUES(?,?);""",(u,p))
        msg3 = msgbox.showinfo( "Status","successfully registered")
        print("List of all users : ")
        cur=conn.execute("SELECT * FROM USER")
        for row in cur:
            print(row)
        print()
        
    else:#else alert the user 
        msg3 = msgbox.showinfo( "Alert","User already exists !!!")
    
    db.commit()
    db.close()
    

#register button
reg=Button(tab2,text="Register",font=('Merriweather',13),command=register)
reg.place(relx=0.72,rely=0.475,anchor=CENTER)
reg.config(bg="#f4d160")
 
#mainloop
root.mainloop()
```
